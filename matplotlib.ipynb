### Two ways of plotting with matplotlib
- using fig class gives full control

```python
fig, ax = plt.subplots(2)    #create figure
ax[0].plot([x for x in range(25)],[y**2 for y in range(25)]) #first subplot
ax[1].plot([x for x in range(25)],[y**3 for y in range(25)]) #second subplot
plt.show() #plot everything
```

- using pyplot Api is less flexible however easier/faster 

```python
plt.subplot(2,1,1)  #rows, columns, using first plot

plt.plot([x for x in range(25)],[y**2 for y in range(25)])  #lists of x values,  y values
plt.plot([x for x in range(25)],[y for y in range(25)])     #second graph
plt.title('Square Numbers')

plt.subplot(2,1,2)  #working in second plot
plt.plot([x for x in range(25)],[y**3 for y in range(25)])
plt.title('Cubed Numbers')

plt.subplots_adjust(hspace=0.5)  #adjust the spacing between subplots
plt.show()
```

```python
import numpy as np

x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')     #creates axes
plt.plot(x, x**2, label='quadratic')  #adds to the current axes
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()

plt.show()
```

## 2D plot

```python
features = iris.data.T
Labels = ['Setosa','Versicolour','Virginica']
colors = ['r','g','b']
classlbl = iris.target


fig = plt.figure(figsize = (12,10))
ax = fig.add_subplot(111)

colors = np.asarray(colors)
colorslist = colors[classlbl]
Labels = np.asarray(Labels)
labellist = Labels[classlbl]

for x,y,s,c,l in zip(features[0],features[1],features[3],colorslist,labellist):
    ax.scatter(x,y,s=100*s,color =c,label =l, cmap='viridis',alpha=0.2)  #petal width is represented by the size of circle
    
handles, labels = ax.get_legend_handles_labels()

newLeg = dict()

for h,l in zip(handles,labels):
    if l not in newLeg.keys():
        newLeg[l]=h
        
#create new handles and labels
handles = []
labels = []
for l in newLeg.keys():
    handles.append(newLeg[l])
    labels.append(l)

#create new Legend
ax.legend(handles,labels,loc='lower right',frameon=True,prop={'size':20})
ax.set_xlabel(iris.feature_names[0],fontsize='22')
ax.set_ylabel(iris.feature_names[1],fontsize='22')
ax.set_title('Demonstration of Data Visualisation',fontsize ='22')
plt.show()
```

- Setosa seems easy to clasify
- Versicolour and Virginica need more insight

## 3D plot

```python
from mpl_toolkits.mplot3d import Axes3D
features = iris.data.T
Labels = ['Setosa','Versicolour','Virginica']
colors = ['r','g','b']
classlbl = iris.target

fig = plt.figure(figsize = (12,10))
ax = fig.add_subplot(111,projection='3d')

colors = np.asarray(colors)
colorslist = colors[classlbl]
Labels = np.asarray(Labels)
labellist = Labels[classlbl]

for x,y,z,s,c,l in zip(features[0],features[1],features[2],features[3], colorslist,labellist):
    ax.scatter(x,y,z,s=100*s,color =c,label =l, cmap='viridis',alpha=0.2)

handles, labels = ax.get_legend_handles_labels()

newLeg = dict()

for h,l in zip(handles,labels):
    if l not in newLeg.keys():
        newLeg[l]=h
        
#create new handles and labels
handles = []
labels = []
for l in newLeg.keys():
    handles.append(newLeg[l])
    labels.append(l)

#create new Legend
ax.legend(handles,labels,loc='lower right',frameon=True,prop={'size':20})
ax.set_xlabel(iris.feature_names[0],fontsize='22')
ax.set_ylabel(iris.feature_names[1],fontsize='22')
ax.set_zlabel(iris.feature_names[2],fontsize='22')
ax.set_title('Demonstration of Data Visualisation',fontsize ='22')
plt.show()
```

## Using Pandas for visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets

iris_dataset = datasets.load_iris()
X = iris_dataset.data


colors = ['r','g','b']
classlbl = iris_dataset.target

colors = np.asarray(colors)
Y = colors[classlbl]


iris_dataframe = pd.DataFrame(X, columns = iris_dataset.feature_names)
#create a scatter matrix from the dataframe, color by y
grr = pd.plotting.scatter_matrix(iris_dataframe, c = Y, figsize = (15,15), marker='o', hist_kwds={'bins':20}, s= 60, alpha = .8)
```
